#  ML Modeling

```python
#Making copy of data for model 1
df_m1 = tot_df.copy()
```

```python
#Dropping the limit and overspeed 
df_m1.drop(['LIMIT','OVERSPEED'],axis=1,inplace=True)
```

```python
#Splitting the x and y
x = df_m1.drop(['SPEED'],axis=1)
y = df_m1['SPEED']
```

```python
#Splitting the data into train and test
from sklearn.model_selection import train_test_split
```

```python
#Data is split into train and test variables
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=47,)
```

```python
print("x train: ",x_train.shape)
print("y train: ",y_train.shape)
print("x test: ",x_test.shape)
print("y test: ",y_test.shape)
```

```python
#Speed our target is continuous, so we will import the model related to regression
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,r2_score,mean_squared_error
```

```python
models = [LinearRegression(),DecisionTreeRegressor(),RandomForestRegressor(), GradientBoostingRegressor()]

for model in models:
  mod = model.fit(x_train,y_train)
  pred = mod.predict(x_test)
  
  mae = mean_absolute_error(y_test,pred)
  mse = mean_squared_error(y_test,pred)
  r2 = r2_score(y_test,pred)

  cvs = cross_val_score(mod,x,y,cv=5)
  
  print("Metrics of model: ",mod)
  print("Mean Absolute Error :",mae)
  print("Mean Squared Error :",mse)
  print("r2 score:",r2)
  print("cv score: ",cvs.mean())
  print("******"*6)
```

From the above models we can see gradient boosting is working better than others, mae is less, r2 score should be more so we will try to increase this further by hyper parameter tuning.

# Considering several factors like mean absolute error and r2 score. and the difference of r2 score and cv score is evaluated here.Looking at these factor gradientboostregressor model is deemed as best.

```python
gbr = GradientBoostingRegressor(random_state=150)

gbr_mod = gbr.fit(x_train,y_train)
gbr_pred = gbr_mod.predict(x_test)

gbr_mae = mean_absolute_error(y_test,gbr_pred)
gbr_mse = mean_squared_error(y_test,gbr_pred)
gbr_r2 = r2_score(y_test,gbr_pred)

gbr_cvs = cross_val_score(mod,x,y,cv=5)

print("Mean Absolute Error :",gbr_mae)
print("Mean Squared Error :",gbr_mse)
print("r2 score:",gbr_r2)
print("cv score: ",gbr_cvs.mean())
```

```python
results = pd.DataFrame()
results['original'] = y_test
results['predicted'] = gbr_pred
results['diff'] = np.abs(results['original'] - results['predicted'])
```
